```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies.head(1)
```

```python
all_movies = movies.merge(credits , on='title')
all_movies.head(1)
```

```python
import ast
```

```python
def convert(obj):
    m_list =[]
    for i in ast.literal_eval(obj):
        m_list.append(i['name'])
    return m_list
```

```python
all_movies.iloc[0 , 1]
```

```python
all_movies['all_genres'] = all_movies['genres'].apply(convert)
```

```python
all_movies.insert(3 , "all_genres", all_movies['genres'].apply(convert))
```

```python
all_movies.insert(4 , "all_keywords", all_movies['keywords'].apply(convert))
```

```python
all_movies.head(0)
```

```python
def convert3(obj):
    counter = 0
    m_list =[]
    for i in ast.literal_eval(obj):
        if counter<=3:
          m_list.append(i['name'])
          counter+=1
        else:
           break  
    return m_list
```

```python
all_movies.insert(5 , "top_cast", all_movies['cast'].apply(convert3))
```

```python
all_movies.head(1)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 
```

```python
all_movies.insert(6 , "movie_director", all_movies['crew'].apply(fetch_director))
```

```python
all_movies.head()
```

```python
all_movies['overview'][0]
```

```python
def convert4(obj):
    new_obj = str(obj).split()
    return new_obj
```

```python
all_movies.insert(7 , "m_overview", all_movies['overview'].apply(convert4))
```

```python
all_movies.head()
```

```python
head_list = list(all_movies.head())
head_list
```

```python
movie_df = all_movies.drop(['budget','genres','homepage','keywords',
 'original_language',
 'original_title',
 'overview',
 'popularity',
 'production_companies',
 'production_countries',
 'release_date',
 'revenue',
 'runtime',
 'spoken_languages',
 'status',
 'tagline',
 'vote_average',
 'vote_count',
 'movie_id',
 'cast',
 'crew'], axis=1)
```

```python
movie_df.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movie_df.insert(2,"genres", movie_df['all_genres'].apply(collapse))
```

```python
movie_df.insert(3,"keywords", movie_df['all_keywords'].apply(collapse))
movie_df.insert(4,"cast", movie_df['top_cast'].apply(collapse))
movie_df.insert(5,"director", movie_df['movie_director'].apply(collapse))
```

```python
movie_df.head()
```

```python
movie_df.insert(2 , "tags", movie_df['m_overview'] + movie_df['genres'] + movie_df['keywords']+movie_df['cast']+movie_df['director'])
```

```python
newdf = movie_df[['id','title','tags']]
newdf.head()
```

```python
def converting(obj):
    newobj = " ".join(obj)
    newobj2 = newobj.lower()
    return newobj2
```

```python
abc = ['a','b','c']
converting(abc)
```

```python
newdf.insert(3 , "all_tags" , newdf['tags'].apply(converting))
```

```python
final_df = newdf.drop('tags', axis =1)
```

```python
final_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
final_df['all_tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(final_df['all_tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])
```

```python
def recommend(movie):
    movie_index = final_df[final_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(final_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
